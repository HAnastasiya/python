```python
#Определите какому фильму было выставлено больше всего оценок 5.0.
import pandas as pd
data=pd.read_csv('ratings.csv',sep=',')
data_1=pd.read_csv('movies.csv',sep=',')
data_new=data_1.merge(data, on = 'movieId',how='outer')
b=data_new[data_new['rating']==5.0]
c=b.groupby('title').count()

c.sort_values('rating',ascending=False).head(10)

#Получается, что фильм Побег из Шоушенка самый рейтинговый.
```

```python
#Задание 2
#По данным файла power.csv посчитайте суммарное потребление стран Прибалтики
#(Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. 
#Не учитывайте в расчетах отрицательные значения quantity.

import pandas as pd
data=pd.read_csv('power.csv',sep=',')

filter_country=data[(data['country']=='Latvia')|(data['country']=='Lithuania')|(data['country']=='Estonia')]
#data[data['country'].str.contains('est',case=False)].head(10)
#(a['date']>2005)
filter_year=filter_country[(filter_country['year']<=2010)&(filter_country['year']>=2005)]
filter_cat=filter_year[(filter_year['category']==4)|(filter_year['category']==12)|(filter_year['category']==21)]
filter_min=filter_cat[filter_cat['quantity']>0]
filter_min.groupby('country').sum()
```
